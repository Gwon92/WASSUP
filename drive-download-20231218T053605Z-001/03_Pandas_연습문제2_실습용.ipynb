````python
df_covid = pd.read_csv("C:\workspace\python\WASSUP\data/seoul-covid19_6_30_.csv")
df_covid
````

# Q1-2
```
'연번' 컬럼은 삭제하고  '환자' 컬럼 값 기준으로 정렬한 후 '환자' 컬럼을 인덱스로 설정하여라. 
```

````python
df_covid = df_covid.drop('연번', axis=1)
df_covid = df_covid.sort_values(by='환자')
df_covid = df_covid.set_index("환자")

df_covid
````

# Q1-3
어느 지역에서 제일 많은 확진자가 발생했는지 1~5위 까지 출력하여라 (거주지 기준)

````python
count = df_covid.groupby('거주지').size()
count = count.sort_values(ascending=False)
count.head(5)
````


# Q1-4
확진자가 가장 많이 발생한 1, 2, 3위 지역의 데이타만 출력하여라 

````python

df_covid[df_covid['거주지'].isin(count.head(3).index)]
````

# Q1-5
'확진일' 컬럼값을 아래와 같은 형태로 변경하여라
<br>
'1.24.' => '2020-1-24'

````python
#df_covid['확진일'] = df_covid['확진일'].str.rstrip('.').replace('.', '-').apply(lambda x: '2020-' + x)
df_covid['확진일'] = df_covid['확진일'].str.replace('.', '-')
df_covid
````

# Q1-6
'확진일' 컬럼을 시계열 데이타로 형태를 변경하여라 

````python
df_covid['확진일'] = pd.to_datetime(df_covid['확진일'], format='%Y-%m-%d')
df_covid
````

# Q1-7
3월에 발생한 확진자의 데이타만 출력하여라.

````python
df_covid[df_covid['확진일'].dt.month == 3]
````

# Q1-8
주말에 발생한 확진자의 데이타만 출력하여라.

````python
df_covid[df_covid['확진일'].dt.weekday.isin([5, 6])]
````

# Q1-9
'콜센터'와 관련된 확진자의 데이타를 출력하여라. 

````python
df_covid[df_covid['접촉력'].str.contains('콜센터')]
````

## Q1-10
'우한'과 관련된 확진자의 데이타만 출력하여라. 

````python
df_covid[df_covid['여행력'].str.contains('우한')]
````
