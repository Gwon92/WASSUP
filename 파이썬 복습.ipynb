```python
cnt = int(input('크롤링 할 건수는 몇건입니까?: '))
page_cnt = math.ceil(cnt / 10) # 크롤링 할 전체 페이지 수 (올림)
print(page_cnt)
```

```python
tit_xpath = '//*[@id="search_result"]/ul/li[*]/div[1]/div[1]/a'
no = 0

for x in range(1, page_cnt+1):
    print(f'========= {x} 페이지 작업 =========')
    mylist = driver.find_elements(By.XPATH, tit_xpath)
    
    for item in mylist:
        no += 1
        if no > cnt:
            break
        print(no, item.text.replace('\n', ''))
    
    if no <= cnt:
        a = f'/html/body/div[3]/div/div[1]/div[15]/a[{x+1}]'
        driver.find_element(By.XPATH, a).click()
        time.sleep(10)
        
print('========= 작업 완료 =========')
driver.close()
```

```python
#스크롤을 최하단으로 내린다.
#상품 제목 값을 찾아 리스트에 저장한다.
#다 찾았으면 다음 페이지로 이동한다.

for i in range(0,2):
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    data = driver.find_elements(By.XPATH, "//*[@id='content']/div[1]/div[2]/div/div[*]/div/div/div[2]/div[1]/a")
    
    for title in data:
        print(title.text)
    
    print("-"*10,"여기까지","-"*10)
    if i == 0:
        driver.find_element(By.XPATH, '//*[@id="content"]/div[1]/div[4]/div/a[1]').click()
    time.sleep(5)
    
```
